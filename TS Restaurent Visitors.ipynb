```python
# sequential sampling
rv1_train = rv1.iloc[0:450]
rv1_test = rv1.iloc[450::]
```

```python
# lets build a SARIMAx models
from pmdarima import auto_arima
```

```python
auto_arima(rv1.total, seasonal=True, m=7, exogenous = rv1[['holiday']])
# taking holiday column as exogenous factor
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model_sarimax_rv = SARIMAX(rv1_train.total, order = (1,0,0), seasonal_order = (1,0,1,7), exog = rv1_train[['holiday']])
# note in auto arima we are using exogenous =  and in sarima we are using exog
```

```python
result_sarimax = model_sarimax_rv.fit()
```

```python
pred_sarimax = result_sarimax.predict(start = 450, end = 477, exog = rv1_test[['holiday']])
```

```python
err_sarimax = rv1_test.total - pred_sarimax
```

```python
mse_sarimax = np.mean(np.square(err_sarimax))
mse_sarimax
```

```python
mape_sarimax = np.mean(np.abs(err_sarimax*100/rv1_test.total))
mape_sarimax
```
